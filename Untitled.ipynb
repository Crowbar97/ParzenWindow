```python
import numpy as np
from scipy.special import gamma
from operator import add
from math import sqrt, pi, exp
from functools import reduce
import matplotlib.pyplot as plt
import matplotlib.pylab as pylab
params = {'legend.fontsize': 'x-large',
          'figure.figsize': (15, 5),
          'axes.labelsize': 'x-large',
          'axes.titlesize': 'x-large',
          'xtick.labelsize': 'x-large',
          'ytick.labelsize': 'x-large'}
pylab.rcParams.update(params)
```

```python
class Distributions:
    @staticmethod
    def normal_density(x, mean, variance):
        return 1 / sqrt(2 * pi * variance ** 2) * exp(- (x - mean) ** 2 / (2 * variance ** 2))

    @staticmethod
    def normal(mean, variance, count):
        return np.random.normal(mean, variance, int(count))

    @staticmethod
    def uniform_density(x, left_bound, right_bound):
        if left_bound <= x <= right_bound:
            return 1 / (right_bound - left_bound)
        return 0

    @staticmethod
    def uniform(left_bound, right_bound, count):
        return np.random.uniform(left_bound, right_bound, int(count))

    @staticmethod
    def beta_density(x, a, b):
        return gamma(a + b) / (gamma(a) * gamma(b)) * x ** (a - 1) * (1 - x) ** (b - 1)

    @staticmethod
    def beta(a, b, count):
        return np.random.beta(a, b, int(count))
```

```python
class Parzen:
    @staticmethod
    def rect_kernel(z):
        return 0.5 * int(abs(z) < 1)

    @staticmethod
    def density(x, points, h):
        sum = 0
        for p in points:
            sum += Parzen.rect_kernel((x - p) / h)
        return 1 / (len(points) * h) * sum

    @staticmethod
    def plot(distr_name, distr_points,
             check_points,
             distr_density, est_density,
             point_count, h):
        plt.figure()

        plt.plot(distr_points, [0] * len(distr_points), 'g|', label=distr_name + " distr")
        plt.plot(check_points, distr_density, 'go-', label=distr_name + " density")
        plt.stem(check_points, est_density, '-.', basefmt=" ", label="est density")

        plt.title("Parzen window density estimation"
                  "\npoint count = " + str(point_count)
                  + "\nbest h = " + str(h), fontsize=20)
        plt.xlabel("x", fontsize=20)
        plt.ylabel("density", fontsize=20)
        plt.grid(True)
        plt.legend(loc=0, fontsize=20)

        plt.show()
        pass
    
    @staticmethod
    def estimate_best_h(distr_name, distr_density,
                        distr_points, check_points,
                        h_min=0.01, h_max=1, h_step=0.01):
        min_error = 1e9
        best_h = 0
        best_density = []
        print("Estimation", distr_name, "distribution density...")
        for h in np.arange(h_min, h_max, h_step):
            est_density = list(map(lambda p: Parzen.density(p, distr_points, h), check_points))
            mse = LossFunctions.mean_square_error(distr_density, est_density)
            print("h =", round(h, 2), ": mse =", mse)
            if mse < min_error:
                min_error = mse
                best_h = h
                best_density = est_density

        print("best h =", round(best_h, 2), ": mse =", min_error)

        Parzen.plot(distr_name,
                    distr_points,
                    check_points,
                    distr_density, best_density,
                    len(distr_points), round(best_h, 2))
```

```python
class LossFunctions:
    @staticmethod
    def mean_square_error(d1, d2):
        return reduce(add, map(lambda y1, y2: (y1 - y2) ** 2, d1, d2)) / len(d1)
```

```python
# conditions
point_count = 1e3
check_points = np.arange(-4, 4, 0.1)

# normal distribution
mean, variance = 0, 1
normal_density = list(map(lambda p: Distributions.normal_density(p, mean, variance), check_points))
normal_points = Distributions.normal(mean, variance, point_count)
Parzen.estimate_best_h("normal", normal_density,
                       normal_points, check_points)
```

```python
 # uniform  distribution
left_bound = -2
right_bound = 2
uniform_density = list(map(lambda p: Distributions.uniform_density(p, left_bound, right_bound), check_points))
uniform_points = Distributions.uniform(left_bound, right_bound, point_count)
Parzen.estimate_best_h("uniform", uniform_density,
                       uniform_points, check_points)
```

```python
 # beta  distribution
check_points = np.arange(0.01, 1, 0.01)
a = 0.5
b = 0.5
beta_density = list(map(lambda p: Distributions.beta_density(p, a, b), check_points))
beta_points = Distributions.beta(a, b, point_count)
Parzen.estimate_best_h("beta", beta_density,
                       beta_points, check_points)
```
